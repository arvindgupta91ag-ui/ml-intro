3.  Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.
- 🔑 Overfitting
Definition: When a model learns the training data too well, including noise and random fluctuations, but fails to generalize to new/unseen data.

Symptoms: Very high accuracy on training data, poor accuracy on test data.

Example: A decision tree that grows too deep and memorizes every training point but misclassifies new inputs.

Solution: Use regularization, reduce model complexity, or apply cross‑validation.
🔑 Underfitting
Definition: When a model is too simple to capture the underlying patterns in the data.

Symptoms: Poor accuracy on both training and test data.

Example: Using a straight line (linear regression) to fit data that clearly follows a curved trend.

Solution: Increase model complexity, add more features, or reduce bias in assumptions.
Bias‑Variance Tradeoff
Bias: Error due to overly simplistic assumptions in the model (leads to underfitting).

Variance: Error due to excessive sensitivity to training data (leads to overfitting).

Tradeoff:

High bias → model is too rigid, misses patterns.

High variance → model is too flexible, captures noise.

The goal is to find a balance where the model generalizes well.

Example:

High bias: Linear regression on complex data.

High variance: Deep neural network trained on small dataset.
| Concept | Definition | Symptom | Example | Fix |
| --- | --- | --- | --- | --- |
| **Overfitting** | Learns noise, poor generalization | High train accuracy, low test accuracy | Deep decision tree | Regularization, pruning |
| **Underfitting** | Too simple, misses patterns | Low train & test accuracy | Linear regression on curved data | More features, complex model |
| **Bias‑Variance Tradeoff** | Balance between simplicity & flexibility | Tradeoff between bias and variance | Linear vs deep models | Cross‑validation, tuning |

4. What are outliers in a dataset, and list three common techniques for
handling them.
- Outliers are data points that differ significantly from the majority of observations in a dataset. They can occur due to variability in measurement, data entry errors, or genuine rare events. Outliers can distort statistical analyses, affect model accuracy, and mislead conclusions if not handled properly.
Common Techniques for Handling Outliers
Removal (Deleting Outliers)

If outliers are due to errors or irrelevant anomalies, they can be removed.

Example: Dropping a record where a person’s age is entered as 250 years.

Transformation (Scaling/Log Transformation)

Applying mathematical transformations (like log, square root, or normalization) reduces the impact of extreme values.

Example: Using log transformation on income data to minimize the effect of very high salaries.
Imputation or Capping (Winsorization)

Replace extreme values with more reasonable ones, such as the nearest percentile value.

Example: Replacing values above the 95th percentile with the 95th percentile itself.
| Technique | Approach | Example Use Case |
| --- | --- | --- |
| **Removal** | Delete extreme values | Age recorded as 250 years |
| **Transformation** | Apply log/sqrt scaling | Income distribution |
| **Imputation/Capping** | Replace with percentile values | Limiting house prices above 99th percentile |

5. Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.
- Handling missing values is a crucial step in data preprocessing because incomplete data can bias results or reduce model accuracy. The process typically involves:
Steps in Handling Missing Values
Identify Missing Data

Use functions like isnull() or info() in Python (Pandas) to detect missing entries.

Analyze the Pattern

Check if missing values are random or systematic.

Random missingness can often be imputed; systematic missingness may need deeper investigation.

Choose a Strategy

Deletion: Remove rows/columns with too many missing values.

Imputation: Fill missing values with estimated values.

Model-based methods: Predict
Imputation Techniques
For Numerical Data:

Mean/Median Imputation → Replace missing values with the mean or median of the column.

Example: If some customers’ ages are missing, fill them with the median age of all customers.

For Categorical Data:

Mode Imputation → Replace missing values with the most frequent category.

Example: If “Gender” has missing entries, fill them with the most common category (e.g., “Male” or “Female”).
| Data Type | Technique | Example |
| --- | --- | --- |
| **Numerical** | Mean/Median Imputation | Fill missing ages with median age |
| **Categorical** | Mode Imputation | Fill missing gender with most frequent value |

8. : Write a Python script to:
● Generate 1000 samples from a normal distribution.
● Introduce 50 random missing values.
● Fill missing values with the column mean.
● Plot a histogram before and after imputation.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(len(data), size=50, replace=False)
data_with_nans = data.copy()
data_with_nans[missing_indices] = np.nan

# Step 3: Create DataFrame
df = pd.DataFrame({'Value': data_with_nans})

# Step 4: Plot histogram before imputation
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Before Imputation')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Step 5: Fill missing values with column mean
mean_value = df['Value'].mean()
df['Value_filled'] = df['Value'].fillna(mean_value)

# Step 6: Plot histogram after imputation
axes[1].hist(df['Value_filled'], bins=30, color='salmon', edgecolor='black')
axes[1].set_title('After Imputation')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

10. 10: You are working as a data scientist for a retail company. You receive a customer
transaction dataset that contains:
● Missing ages,
● Outliers in transaction amount,
● A highly imbalanced target (fraud vs. non-fraud),
● Categorical variables like payment method.
Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.
(Include your Python code and output in the code box below.)
- Step‑by‑Step Plan
Handle Missing Ages

Detect missing values.

Impute using median (robust to outliers) or predictive models.

Handle Outliers in Transaction Amount

Detect using IQR (Interquartile Range) or Z‑score.

Options: remove, cap (Winsorization), or transform (log scaling).

Handle Imbalanced Target (Fraud vs Non‑Fraud)

Techniques:

Resampling (SMOTE for oversampling minority, undersampling majority).
Class weights in ML models.

Encode Categorical Variables (Payment Method)

Use One‑Hot Encoding for nominal categories.

Use Label Encoding if categories are ordinal.

Final Step

Scale numerical features (e.g., Min‑Max or StandardScaler).

Split into train/test sets.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# Example synthetic dataset
data = {
    'Age': [25, np.nan, 40, 35, np.nan, 50, 28],
    'TransactionAmount': [100, 2000, 150, 120, 50000, 300, 80],
    'PaymentMethod': ['Card', 'Cash', 'Card', 'UPI', 'Cash', 'UPI', 'Card'],
    'Fraud': [0, 0, 0, 0, 1, 0, 0]
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# 1. Handle Missing Ages (median imputation)
imputer = SimpleImputer(strategy='median')
df['Age'] = imputer.fit_transform(df[['Age']])

# 2. Handle Outliers (cap using IQR method)
Q1 = df['TransactionAmount'].quantile(0.25)
Q3 = df['TransactionAmount'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
df['TransactionAmount'] = np.where(df['TransactionAmount'] > upper_bound,
                                   upper_bound,
                                   np.where(df['TransactionAmount'] < lower_bound,
                                            lower_bound,
                                            df['TransactionAmount']))

# 3. Handle Imbalance (SMOTE oversampling)
X = df.drop('Fraud', axis=1)
y = df['Fraud']

# Encode categorical before SMOTE
encoder = OneHotEncoder(sparse_output=False, drop='first')
X_encoded = encoder.fit_transform(df[['PaymentMethod']])
X_final = np.hstack([df[['Age','TransactionAmount']], X_encoded])

smote = SMOTE(random_state=42)

minority_class_count = y.value_counts().min()
# SMOTE requires at least k_neighbors + 1 samples in the minority class for its default setting (k_neighbors=5).
# If k_neighbors is set to k, it requires k+1 samples.
# With only 1 minority sample, SMOTE is not applicable with default or any meaningful k_neighbors >= 1.

if minority_class_count >= (smote.k_neighbors + 1):
    X_resampled, y_resampled = smote.fit_resample(X_final, y)
    print("\nResampled class distribution:", pd.Series(y_resampled).value_counts())
else:
    print(f"\nSkipping SMOTE: Minority class has only {minority_class_count} sample(s), which is less than required (k_neighbors + 1 = {smote.k_neighbors + 1}).")
    print("SMOTE is not applicable with a minority class size of 1. Proceeding without oversampling.")
    X_resampled, y_resampled = X_final, y # Proceed without SMOTE
    print("\nClass distribution after skipping SMOTE:", pd.Series(y_resampled).value_counts())

# 4. Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

print("\nPrepared feature matrix shape:", X_scaled.shape)
```
